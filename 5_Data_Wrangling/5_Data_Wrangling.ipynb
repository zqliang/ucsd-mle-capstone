# Data Wrangling

## Load data

```python
import pandas as pd
import numpy as np

df_anime = pd.read_csv('../data/anime-dataset-2023.csv')
print(df_anime.shape)
df_anime.head()
```

```python
df_users = pd.read_csv('../data/users-details-2023.csv')
print(df_users.shape)
df_users.head()
```

```python
df_ratings = pd.read_csv('../data/users-score-2023.csv')
print(df_ratings.shape)
df_ratings.head()
```

### Remove duplicates

```python
#find duplicates

duplicates_all = df_anime[df_anime.duplicated()]
print("All Duplicates:")
print(len(duplicates_all))

duplicates = df_anime[df_anime.duplicated(['Name'])].sort_values(by='Name')
print("Duplicates based on Name:")
print(len(duplicates))
duplicates = duplicates[['anime_id', 'Name']]
print(duplicates)

df_anime_new = df_anime.drop_duplicates(['Name'])
print("Cleaned anime shape: {} \n".format(df_anime_new.shape))
print("Old anime shape: {}".format(df_anime.shape))
```

```python
duplicates_all = df_users[df_users.duplicated()]
print("All Duplicates across users:")
print(len(duplicates_all))

duplicates = df_users[df_users.duplicated(['Username'])]
print("\nDuplicates Based on Username:")
print(len(duplicates))
```

```python
duplicates = df_ratings[df_ratings.duplicated(['Username', 'Anime Title'])]
print("\nDuplicates Based on Username and Anime Title:")
print(len(duplicates))
print(duplicates)

df_ratings_new = df_ratings.drop_duplicates(['Username', 'Anime Title'])
print("Cleaned scores shape: {} \n".format(df_ratings_new.shape))
print("Old scores shape: {}".format(df_ratings.shape))
```

### Visualize data

```python
import seaborn as sns
import matplotlib.pyplot as plt
import re
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

df_anime_years = df_anime_new.copy()

df_anime_years['Aired'] = df_anime_years['Aired'].astype(str)

#look for the release year in Aired column
for idx in df_anime_years.index:
    aired_value = df_anime_years.at[idx, 'Aired']
    match = re.search(r'\b(\d{4})\b', aired_value)
    
    if match:
        release_year = match.group(0)
        df_anime_years.loc[idx, 'Release Year'] = release_year
    else:
        df_anime_years.loc[idx, 'Release Year'] = 'unknown'

        
#display the # of animes released over the years
df_anime_years = df_anime_years.loc[df_anime_years['Release Year'] != 'unknown']
    
fig, ax = plt.subplots(1, 1, figsize=(14, 6))


data = df_anime_years['Release Year'].value_counts().sort_index()
x = data.index.map(int)
y = data.values

sns.barplot(x=x, y=y) #barplot expects x and y to be passed as named arguments
xmin, xmax = plt.xlim()
xtick_labels = [x[0]] + list(x[5:-5:5]) + [x[-1]]
plt.xticks(ticks=np.linspace(xmin, xmax, len(xtick_labels)), labels=xtick_labels)
#includes the first year, a selection of years at regular intervals 
#excludes first and last 5 years, and the last year

plt.ylabel("# of anime")
plt.title("Release year distribution")
plt.show()
```

```python
ratings = df_ratings['rating'].value_counts().sort_index()
print(ratings)

custom_palette = sns.color_palette("flare")
sns.barplot(x=x, y=ratings.values, palette=custom_palette)

#format y labels to show in millions
plt.ticklabel_format(style='plain', axis='y')
plt.gca().yaxis.get_major_formatter().set_scientific(False)
plt.tick_params(axis='y', labelrotation=0)
plt.ylabel("# of ratings")

plt.title("User rating distribution")
plt.show()
```

```python
# top 10 most popular anime
df_anime_new['Popularity'] = pd.to_numeric(df_anime_new['Popularity'], errors='coerce')
top_10_popular = df_anime_new.sort_values(by='Popularity').head(10)


sns.barplot(x='Name', y='Popularity', data=top_10_popular, palette='viridis')
plt.title('Top 10 Most Popular Anime')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')  #rotate x-axis labels for better visibility
plt.show()
```

```python
selected_columns = ['Episodes', 'Score', 'Members', 'Popularity', 'Rank']

sns.pairplot(df_anime_new[selected_columns])
plt.suptitle('Pair Plot of Anime Features', y=1.02)
plt.show()

plt.figure(figsize=(8, 6))
sns.heatmap(df_anime_new[selected_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Anime Features')
plt.show()
```

```python
selected_columns = ['Days Watched', 'Episodes Watched', 'Mean Score', 'Completed', 'Dropped', 'Rewatched']


plt.figure(figsize=(8, 6))
sns.heatmap(df_users[selected_columns].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of User Features')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Based on the heatmaps, rank and popularity are strongly positively correlated. We see that there is a strong negative correlation between
rank and score and also between popularity and score. Anime with popularity=1 is most popular, meaning a less popular anime
will have lower score. There is a slightly positive correlation between members and score, so higher scored anime tend to have a larger audience. There is no significant correlation between user watching behavior and score.
